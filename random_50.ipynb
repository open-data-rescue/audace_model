```python
# Shuffle the DataFrame randomly
shuffled_df = lapatrie_samples['Date'].sample(frac=1, random_state=42)

# Calculate the number of rows per person
num_rows = len(shuffled_df)
rows_per_person = num_rows // 3

# Assign the rows to each person
person1 = shuffled_df[:rows_per_person]
person2 = shuffled_df[rows_per_person:2*rows_per_person]
person3 = shuffled_df[2*rows_per_person:]

# Print the assigned data for each person
print("Person 1:")
print(person1)
print()
print("Person 2:")
print(person2)
print()
print("Person 3:")
print(person3)
```

```python
# Shuffle the DataFrame randomly
shuffled_df = lapress_samples['Date'].sample(frac=1, random_state=42)

# Calculate the number of rows per person
num_rows = len(shuffled_df)
rows_per_person = num_rows // 3

# Assign the rows to each person
person1 = shuffled_df[:rows_per_person]
person2 = shuffled_df[rows_per_person:2*rows_per_person]
person3 = shuffled_df[2*rows_per_person:]

# Print the assigned data for each person
print("Person 1:")
print(person1)
print()
print("Person 2:")
print(person2)
print()
print("Person 3:")
print(person3)
```
